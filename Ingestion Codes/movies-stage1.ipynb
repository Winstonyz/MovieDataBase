```python
df_movies = pandas.read_csv('./title.basics.csv')
```

```python
df_movies.head()
```

```python
df_movies.info()
```

```python
df_movies_60P = df_movies [(df_movies.runtimeMinutes >= 60)]
```

```python
df_movies_60P_movie = df_movies_60P [(df_movies_60P.titleType == 'movie')]
```

```python
df_movies_60P_movie.head()
```

```python
df_movies_60P_movie.info()
```

```python
df_movies_60P_movie_start1990 = df_movies_60P_movie [(df_movies_60P_movie.startYear >= 1990)]
```

```python
df_movies_60P_movie_start1990.head();

df_movies_60P_movie_start1990.info();
```

```python
df_movies_60P_movie_start1990 = df_movies_60P_movie_start1990[['tconst', 'primaryTitle', 'isAdult', 'startYear', 'runtimeMinutes', 'genres']]
```

```python
df_movies_60P_movie_start1990.head();

df_movies_60P_movie_start1990.info();
```

```python
df_movies_60P_movie_start1990 = df_movies_60P_movie_start1990 [(df_movies_60P_movie_start1990.startYear <= 2020)]
```

```python
df_movies_60P_movie_start1990.head();

df_movies_60P_movie_start1990.info();
```

```python
df_movies_60P_movie_start1990_edit = df_movies_60P_movie_start1990 [(df_movies_60P_movie_start1990.runtimeMinutes >= 80)]
```

```python
df_movies_60P_movie_start1990_edit.info()
```

```python
df_movies_60P_movie_start1990_nodo = df_movies_60P_movie_start1990_edit [(df_movies_60P_movie_start1990_edit.genres != 'Documentary')]
```

```python
df_movies_60P_movie_start1990_nodo.info()
```

```python
df_movies_60P_movie_start1990_nodo = df_movies_60P_movie_start1990_nodo [(df_movies_60P_movie_start1990_nodo.genres != 'History')]
```

```python
df_movies_60P_movie_start1990_nodo.head(20)
```

```python
df_movies_60P_movie_start1990_nodo.to_csv('movies.csv', index=False)
```
