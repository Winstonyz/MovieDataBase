```python
new_writers_edited = new_writers[['nconst', 'primaryName', 'birthYear', 'deathYear', 'knownForTitles']];

new_writers_edited = new_writers_edited[new_writers_edited['birthYear'].notnull()];

new_writers_edited = new_writers_edited[(new_writers_edited.birthYear != '\\N')];

new_writers_edited = new_writers_edited [(new_writers_edited.knownForTitles != '\\N')];

new_writers_edited = new_writers_edited.join(new_writers_edited['knownForTitles'].str.split(',', expand=True).rename(columns={0:'Title1', 1:'Title2', 2:'Title3', 3:'Title4'}));

new_writers_edited = new_writers_edited[['nconst', 'primaryName', 'birthYear', 'deathYear', 'Title1', 'Title2', 'Title3', 'Title4']];

W_Title1 = new_writers_edited[['nconst', 'primaryName', 'birthYear', 'deathYear', 'Title1']];

W_Title2 = new_writers_edited[['nconst', 'primaryName', 'birthYear', 'deathYear', 'Title2']];

W_Title3 = new_writers_edited[['nconst', 'primaryName', 'birthYear', 'deathYear', 'Title3']];

W_Title4 = new_writers_edited[['nconst', 'primaryName', 'birthYear', 'deathYear', 'Title4']];

W_Title1.rename(columns = {'Title1':'Title'}, inplace = True);

W_Title2.rename(columns = {'Title2':'Title'}, inplace = True);

W_Title3.rename(columns = {'Title3':'Title'}, inplace = True);

W_Title4.rename(columns = {'Title4':'Title'}, inplace = True);

W_Title_edited = pandas.concat([W_Title1, W_Title2, W_Title3, W_Title4], ignore_index=True);

written_by = W_Title_edited[['nconst', 'Title']];
```
